## Functions

```python
import sys
sys.path.append('../pso/python/')
from function import *
from figure import plotfunc
```

```python
funcs = [sphere, rosenbrock, rastrigin, griewank]

for func in funcs:
    plotfunc(func, 5.12)
```

```python
models = []
for func in funcs:
    cso = Cso(20,func,-5.12,5.12,2,100)
    models.append(cso)
```

```python
paths = []
for model in models:
    path = model.get_agents()
    paths.append(path)
```

```python
def plot_fit(hist, func):
    best_fit = []
    for it in hist:
        fits = [func(x) for x in it]
        best_fit.append(min(fits).copy())
        
    plt.plot(best_fit, label = str(func))
```

```python
for i, func in enumerate(funcs):
    plot_fit(paths[i],func)
    plt.legend()
```
